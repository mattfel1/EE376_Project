```python
#Information Transduction Capacity of Noisy Biochemical Signaling Networks
#Implementation for EE376a
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
#Solving for SNR (can back calculate SNR if you know MI/capacity) if make assumption about bush network:
n = 1
I =0.85
SNR_bush = (2**(2*I) -1)/n #Bush model
print SNR_bush
print 0.5*math.log(1+(2.25+2.58),2) #This is the empirically derived information using 2 pathways -- NF-kB alone = 0.92 bits; ATF-2 alone = 0.85 bits. Together = 1.27 bits
print 0.5*math.log(1+SNR_bush,2)
print 0.5*math.log(1+SNR_bush*10)
```

```python
#Bush Network
2.58
2.25
SNR = SNR_bush
n = 100
I = np.zeros(n)
for ni in range(1,n):
    I[ni] = 0.92+(0.5*math.log(1+(ni-1)*(2.25+2.58),2))
    #I[ni] = 0.5*math.log((ni)*(2.25),2)
plt.plot(I)
plt.xscale('log')
print(I[14])
```

```python
#Tree Network
SNR = SNR_bush
Noise = 1
Sig = SNR
n = 100
I = np.zeros(n)
for ni in range(1,n):
    #I[ni] = 0.5*math.log(1.0+(ni*3*SNR)/(1.0+ni*SNR),2)
    I[ni] = 0.5*math.log(1+ni*Sig/(1+ni*Noise),2)
plt.plot(I)
print I[n-1]
```
